```python
# ingresar numeroro entero e indicar si es par o impar
z : int
z = int(input("ingrese un numero: "))
if z % 2 == 0:
    print('el numero ' + str(z) + ' es par')
else:
    print('el numero ' + str(z) + ' es impar')
```
